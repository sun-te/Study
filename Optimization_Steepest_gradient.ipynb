```python
# Initial setup
b = -1
delta = 0.02
cov = 0.001
def f(b):
    return 4*b**2 - 5 * b + 9
# Stock the old value of b
hist = [b, b+delta]
f_old = f(b)
# Set for the inertial uncertainty
uncertainty = 2*delta
while uncertainty >= cov:
    f_new = f(hist[1])
    f_now = f(hist[0])
    i = 0
    while f_new < f_now:
        i+=1
        hist[1] += delta
        hist[0] += delta
        f_new = f(hist[1])
        f_now = f(hist[0])
    uncertainty = 2*delta
    print("uncertainty: {}, iteration: {}".format(uncertainty, i))
    if uncertainty < cov:
        res = [hist[0] - delta] + hist
        res_val = [f(x) for x in res]
        print("a1: {}; a2: {}; a3: {}".format(res[0],res[1], res[2]))
        print(res_val)
        break
    hist[0] -= delta
    delta /= 10
    hist[1] = hist[0] + delta
# print(hist)
```
