## 正则表达式

```python
#正则表达式 regular expression
import re
```

```python
p1="cat"
p2="bird"
st="dog runs to cat"
print(re.search(p1,st))
print(re.search(p2,st))
```

### expression using

```python
#a-z 0-9 A-Z自动匹配
ptn=r"r[a-z]n"
print(re.search(ptn,st))
```

```python
# \d: decimal digit
# \D: non decimal digit
# \s:white space (\t\n\r\f\v)
# \S: non
# \w : [a-zA-Z0-9]
# \b : white space
# \\: \
# . :anything apart from white space
# ^: begin
# $:end 
# ? : may or not
print(re.search(r"r\Dn", st))
print(re.search(r"r.n",st))
print(re.search(r"Mon(day)?","Mondy"))
```

### 多行匹配

```python
string="""
dog runs to cat 
I run to dog
"""
print(re.search(r"^I",string,flags=re.M))  #consider every line as independent
```

### group finding

```python
match=re.search(r"(\d+), Date:(.+)","No. 1023, Date: 25/12/2017")
print(match.group(2))
```

```python
# naming the group
match=re.search(r"(?P<id>\d+), Date:(.+)","No. 1023, Date: 25/12/2017")
print(match.group('id'))
```

### replacement

```python
print(re.sub(r"runs", "catches","dog runs cat"))
```

### split

```python
print(re.split(r"[,;\.\]]","a;e.q.K"))
```

# 爬虫

```python
print(html)
```

```python
# findall
# .*? : all information even there is noun
# re.DOTALL : several lines
res=re.findall(r"<h1>(.*?)</h1>",html,flags=re.DOTALL)
print(res)
```

```python
res=re.findall(r"<p>(.*?)</p>",html,flags=re.DOTALL)
print(res[0])
```

```python
#select all links
res=re.findall(r'href="(.*?)"',html)
print(res)
```

## beautiful soup with CSS

```python
html=urlopen('https://morvanzhou.github.io/static/scraping/list.html').read().decode('utf-8')
soup=BeautifulSoup(html,features='lxml')
```

```python
# li is the tag we want
# to take class into consideration
month=soup.find_all('li',{'class':'month'})
for m in month:
    #屏蔽掉所有的html的代码,只出现所需文字
    print(m.get_text())
print(month)
```

```python
jan=soup.find('ul',{'class':'jan'})
print(jan)
day=jan.find_all('li')
for l in day:
    print(l.get_text())
```
