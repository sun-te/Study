### 什么是内存溢出
内存泄漏（Memory Leak）是指程序中己动态分配的堆内存由于某种原因程序未释放或无法释放，造成系统内存的浪费，导致程序运行速度减慢甚至系统崩溃等严重后果。系统中存在无法回收的内存或使用的内存过多，最终使得程序运行要用到的内存大于能提供的最大内存。

原因：

第一种是由于长期保持某些资源的引用，垃圾回收器无法回收它，从而使该资源不能够及时释放，也称为内存泄露； [2] 
另外一种是当需要保存多个耗用内存过大或当加载单个超大的对象时，该对象的大小超过了当前剩余的可用内存空间。

解决方案：

- 检查代码是否有死循环和过分的递归
- 数据库是否由大量一次性读取
- 大型变量使用完后是否没有释放
- 直接检查内存，使用内存内部直接释放

### Machine Learning
对于捉罪犯和检验员工是否迟到（旷工），机器学习会有什么区别， recall rate分别是什么样的情况？

True Positive(真正，TP)：将正类预测为正类数

True Negative(真负，TN)：将负类预测为负类数

False Positive(假正，FP)：将负类预测为正类数误报 (Type I error)

False Negative(假负，FN)：将正类预测为负类数→漏报 (Type II error)

$$Precision = \frac{TP}{FP + TP} = \frac{预测对的正类}{所有预测为正类的样本数}$$
$$Recall = \frac{TP}{FN + TP}= \frac{预测对的正类}{所有真正的正类}$$
$$Accuracy = \frac{TP  + TN}{ALL}$$

抓罪犯的recall率会高一些，因为会抓很多人也不要放过罪犯，因此所有的P被抓到的概率会大一些

### 贝叶斯天气预测
A和B两个公司，他们预测天气的准确率都是60%，那么他们都预测了明天天气为晴天，明天真正晴天的概率是多少？

设：X事件：A站预报明天下雨Y事件：B站预报明天下雨R事件：明天下雨已知X，Y独立。题目求P（R|X，Y）。由Bayes公式：
$$P（R|X）= P（X，R）/ P（X）= P（X|R）P（R）/ P（X）$$

同理：
$$P（R|Y）= P（Y，R）/ P（Y）= P（X|R）P（R）/ P（Y）$$

得出：

$$P（X|R）= P（R|X）P（X）/ P（R）= 0.9 P（X）/ P（R）P（Y|R）= P（R|Y）P（Y）/ P（R）= 0.8 P（Y）/ P（R）$$

X，Y 是independent，所以：

$$P（X，Y）= P（X）P（Y）P（X，Y|R）= P（X|R）P（Y|R）$$
再次用Bayes公式：$$P（X，Y，R）= P（X，Y|R）P（R）= P（X|R）P（Y|R）P（R）= P（R|X，Y）P（X，Y）$$

所以最终：

$$P（R|X，Y）= P（X|R）P（Y|R）P（R）/ P（X，Y）= 0.72/P（R）$$
